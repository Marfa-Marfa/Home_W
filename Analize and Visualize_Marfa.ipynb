### Создайте одномерный массив со значениями от 10 до 49 и разверните его (первый элемент становится последним). Найдите в нем все четные элементы.

```python
array = np.random.randint(10, 50, size=39)
print(array)

# Разверните его (первый элемент становится последним)
print('\n вариант 1')
array_1 = array[::-1]
print(array_1)

print('\n вариант 2')
array_2 = np.flip(array, 0) # 0 - ось
print(array_2)
```

```python
# Найдите в нем все четные элементы
# Метод понимания списка 
even_nos = [num for num in array_1 if num % 2 == 0]
print(even_nos)
```

### Создайте случайный двумерный массив 7x7, найти у него среднее и стандартное оклонение. Нормализуйте этот массив.

```python
array = np.random.randint(1, 10, size=(7,7))
print(array)

# Найти у него среднее:
aver = array.mean()
print(f'\nСреднее = {aver}')

# Найти у него  стандартное оклонение:
skv = np.nanstd(array)
print(f'Стандартное отклонение = {skv}')
```

```python
# Нормализовать массив:

#Для матрицы общая нормализация заключается в использовании евклидовой нормы или нормы Фробениуса.
#Формула для простой нормализации такова: v-cap = v / |v |
#где v - это матрица, а |v | - определитель, который также называется евклидовой нормой. v-cap - это нормализованная матрица.

norm_m = np.linalg.norm(array)
b=(array/norm_m).round(2)
print(b)
```
